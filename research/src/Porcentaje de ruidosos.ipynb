```python
import pandas as pd # se importa pandas como pd
import numpy as np  #numpy como np
import matplotlib.pyplot as plt
```

```python
datos = pd.read_csv('solo_presiones.csv')
```

```python
def ruido(x):
    """
    Determina el tipo de dato ruidoso
    2: menor que 0
    1: menor que 5 (estimación visual)
    0: mayor que 5 (se estima que correcto)
    """
    if x < 0:
        return 2
    elif x < 5:
        return 1
    else:
        return 0
```

```python
p1 = datos.P1.apply(ruido)
p2 = datos.P2.apply(ruido)
p3 = datos.P3.apply(ruido)
p4 = datos.P4.apply(ruido)
p5 = datos.P5.apply(ruido)
p6 = datos.P6.apply(ruido)
```

```python
#Conversión a porcentaje
p1 = p1.value_counts(normalize=True) * 100
p2 = p2.value_counts(normalize=True) * 100
p3 = p3.value_counts(normalize=True) * 100
p4 = p4.value_counts(normalize=True) * 100
p5 = p5.value_counts(normalize=True) * 100
p6 = p6.value_counts(normalize=True) * 100
```

```python
#Ordenación
p1 = p1.sort_index()
p2 = p2.sort_index()
p3 = p3.sort_index()
p4 = p4.sort_index()
p5 = p5.sort_index()
p6 = p6.sort_index()
```
